**Testes Estatísticos entre as Classes**

Com o objetivo de avaliar se as distribuições das variáveis diferem estatisticamente entre transações legítimas e fraudulentas, foi aplicado o **teste de Mann–Whitney U** para cada feature numérica.

Para cada variável, as amostras das classes 0 (legítima) e 1 (fraudulenta) foram comparadas individualmente.

Foi adotado um nível de significância de **α = 0.01**, tornando o critério mais rigoroso e reduzindo o risco de falsos positivos.

Os resultados indicam que a grande maioria das variáveis apresenta **diferenças estatisticamente significativas** entre as classes. Apenas algumas features (`V13`, `V15`, `V22` e `V25`) não apresentaram evidência estatística suficiente para rejeitar a hipótese nula ao nível de significância adotado.


```python
features = df.columns.drop("Class").to_list()

coluna_target = "Class"

analise_mw = {}

classes = df[coluna_target].unique()

alfa = 0.01

agrupamento = df.groupby(coluna_target)
```

```python
for coluna in features:
    grupo0 = agrupamento.get_group(classes[0])[coluna]
    grupo1 = agrupamento.get_group(classes[1])[coluna]

    estatistica_mw, pvalue_mw = mannwhitneyu(grupo0, grupo1)

    analise_mw[coluna] = {
        "pvalue": pvalue_mw,
        "estatistica": estatistica_mw,
        "relacao_sign": pvalue_mw < alfa,
    }

df_mw = pd.DataFrame(analise_mw).T

df_mw
```

**Tratamento de Outliers e Preparação da Base**

Com base nas análises visuais e estatísticas realizadas anteriormente, foi adotada uma estratégia conservadora para o tratamento de outliers.

O dataset foi inicialmente separado entre:
- Transações legítimas (`Class = 0`)
- Transações fraudulentas (`Class = 1`)

Considerando que comportamentos atípicos podem ser justamente os sinais mais relevantes de fraude, optou-se por **não remover outliers da classe fraudulenta**.

Para a classe legítima, foi aplicada uma filtragem baseada em quantis (5% e 95%) para reduzir o impacto de valores extremos que poderiam introduzir ruído no treinamento dos modelos.

Antes do tratamento:
- Classe legítima: 284.315 registros
- Classe fraudulenta: 492 registros

Após a remoção de outliers da classe legítima:
- Classe legítima: 256.088 registros
- Classe fraudulenta: 492 registros

A remoção de outliers foi aplicada exclusivamente à classe majoritária e teve como objetivo reduzir ruído estatístico, sem comprometer a representatividade dos eventos fraudulentos.

```python
df_legitima = df[df["Class"] == 0]
df_fraudulenta = df[df["Class"] == 1]

df_legitima.shape, df_fraudulenta.shape
```

```python
def rem_outliers_quantil(df, variaveis, quantil_inferior=0.05, quantil_superior=0.95):
    df_filtrado = df.copy()
    for variavel in variaveis:
        limite_inferior = df[variavel].quantile(quantil_inferior)
        limite_superior = df[variavel].quantile(quantil_superior)
        df_filtrado = df[(df[variavel] >= limite_inferior) & (df[variavel] <= limite_superior)]
    return df_filtrado
```

```python
variaveis = df.columns.drop("Class")

df_legitima_filtrado = rem_outliers_quantil(df_legitima, variaveis)

df_legitima_filtrado.shape
```

```python
df_sem_outliers = pd.concat([df_legitima_filtrado, df_fraudulenta])

df_sem_outliers.shape
```

```python
df_sem_outliers = df_sem_outliers.reset_index(drop=True)

df_sem_outliers.head()
```
